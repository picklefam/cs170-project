```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_erdos_renyi(n, p):
    G = nx.erdos_renyi_graph(n, p)
    while not nx.is_k_edge_connected(G, 1):
        G = nx.erdos_renyi_graph(n, p)
    return G

def graph_to_txt(G):
    assert(nx.is_k_edge_connected(G, 1))
    n = G.number_of_nodes()
    filename = '%d.in' % n
    graph = open(filename,"w")
    graph.write(str(n) + '\n')
    for u, v in list(G.edges):
        graph.write('%d %d %.3f\n' % (u, v, np.random.rand()*100))
    graph.close()
```

```python
def generate_in(n, p, G=nx.Graph()):
    G = nx.union(G, generate_erdos_renyi(n, p))
    graph_to_txt(G)
    plt.figure()
    nx.draw(G)
```

```python
generate_in(25, 0.15)
```

```python
generate_in(50, 0.05)
```

```python
generate_in(100, 0.03)
```
